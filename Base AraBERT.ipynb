```python
result, texts = model.predict(Test_lista, n_best_size=20)



Not_found=0
string_result=""
string_result+="{"
fpid=False

lenAP=0
for x in range(len(result)):
  line1=result[x]
  line2=texts[x]
  id1=line1['id']
  pid=Dic_id_pid[int(id1)]
  pid=pid.replace(",","")
  ptext=Dic_id_ptext[int(id1)]

  answer1=line1['answer']
  probability1=line2['probability']

  answers_list=[]
  AC=1
  dic_text=""
  f222=False

  Firs_Score=""
  len_Frequence=5-len(answer1)


  for m in range(len(answer1)):
    dica={}
    ans=answer1[m]
    score=probability1[m]
    if ans != "empty":
      SplitAns=str(ans).strip().split()
      SplitPassage=str(ptext).strip().split()

      Found=False
      Start2=0
      End2=0

      sub=str(ans).strip().replace("\"","")
      full=ptext
      ans=full.index(sub)
      Start1=ans
      End1=ans+len(sub)
      string_space=full[:Start1]
      Number_of_space=string_space.count(" ")
      full_list=full.split()
      ans_list=sub.split()
      Start2=Number_of_space
      End2=Number_of_space+len(ans_list)



      ST=Start2
      ET=End2

      newans=SplitPassage[ST:ET]
      String_newans=' '.join(newans)

      if ST==0 and ET==0:
        print(pid)
        print(ans)
        print(ptext)
        print(SplitAns)
        print(SplitPassage)
        print("==============================================================")
        Not_found+=1

      emptyFlage=False
      if ST==ET:
        emptyFlage=True

      if emptyFlage==False and String_newans!=".":
        if f222==False:
          dic_text+="{\"answer\":"+ "\""+str(String_newans).strip()+"\"" + ",\"rank\":"+ str(AC)+ ",\"score\":"+ str(score)+",\"strt_token_indx\":"+ str(ST)+",\"end_token_indx\":"+ str(ET-1) +"}"
          f222=True
        else:
          dic_text+=",{\"answer\":"+ "\""+str(String_newans).strip()+"\"" + ",\"rank\":"+ str(AC)+ ",\"score\":"+ str(score)+",\"strt_token_indx\":"+ str(ST)+",\"end_token_indx\":"+ str(ET-1) +"}"
        AC+=1
        answers_list.append("1")



  if len(answers_list)>0:
    if fpid==False:
      string_result+=pid+":"+"["+dic_text+"]"+"\n"
      fpid=True
    else:
      string_result+=","+pid+":"+"["+dic_text+"]"+"\n"
  else:
    if fpid==False:
      string_result+=pid+": []"+"\n"
      fpid=True
    else:
      string_result+=","+pid+": []"+"\n"




string_result=string_result.strip()+"}"
save_path="/content/sample_data/output/largearab_run20.json"
with open(save_path, 'w', encoding="utf-8") as f:
  f.write(string_result)

```
