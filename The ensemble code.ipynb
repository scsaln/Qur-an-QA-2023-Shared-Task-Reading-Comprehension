```python
import json

import pandas as pd
import arabicstopwords.arabicstopwords as stp

ALL_STRING_REsult=""
ALL_STRING_REsult+="{"

example1 = "/content/sample_data/basearab_run20.json"

file1 = open(example1, "r")
FileContent = file1.read()
file1.close()

FileContent=FileContent[:len(FileContent)-2]
lines1=FileContent.split("\n")

example2 = "/content/sample_data/largearab_run20.json"

file2 = open(example2, "r")
FileContent2 = file2.read()
file2.close()

FileContent2=FileContent2[:len(FileContent2)-2]
lines2=FileContent2.split("\n")




All=0
counter1=0
answers_list=[]
fpid=False

for line_index in range (len(lines1)):
  All=0
  answers_list=[]
  df = pd.DataFrame(columns = ['answer', 'score', 'strt_token_indx', 'end_token_indx'])
  split_tex1=lines1[line_index].split("[")



  part1=split_tex1[0]
  indexes_id = [match.start() for match in re.finditer(r'\"', part1)]
  first_index=indexes_id[0]
  Second_index=indexes_id[1]+1
  pid1=str(part1[first_index:Second_index]).strip()


  part2=","+split_tex1[1]
  part2_split=part2.split("}")



  for index_ans in range(len(part2_split)-1):
    fullans=part2_split[index_ans]
    indexes_twodot = [match.start() for match in re.finditer(r':', fullans)]
    indexes_colun = [match.start() for match in re.finditer(r',', fullans)]
    answer=str(fullans[indexes_twodot[0]+1:indexes_colun[1]]).strip()
    score=float(str(fullans[indexes_twodot[2]+1:indexes_colun[3]]).strip())
    strt_token=str(fullans[indexes_twodot[3]+1:indexes_colun[4]]).strip()
    End_token=str(fullans[indexes_twodot[4]+1:]).strip()
    answers_list.append("1")

    list_row = [answer, score, strt_token , End_token]
    df.loc[len(df)] = list_row


  df


  # second file
  split_tex2=lines2[line_index].split("[")


  part2=","+split_tex2[1]
  part2_split=part2.split("}")

  for index_ans in range(len(part2_split)-1):
    fullans=part2_split[index_ans]
    indexes_twodot = [match.start() for match in re.finditer(r':', fullans)]
    indexes_colun = [match.start() for match in re.finditer(r',', fullans)]
    answer=str(fullans[indexes_twodot[0]+1:indexes_colun[1]]).strip()
    score=float(str(fullans[indexes_twodot[2]+1:indexes_colun[3]]).strip())
    strt_token=str(fullans[indexes_twodot[3]+1:indexes_colun[4]]).strip()
    End_token=str(fullans[indexes_twodot[4]+1:]).strip()
    answers_list.append("1")

    if str(answer).strip() in str(df['answer'].values).strip():
      dindex=df.index[df['answer']==answer].tolist()
      tempdf=df.iloc[dindex[0]]
      df.at[dindex[0],'score']=score+tempdf['score']

    else:
      list_row = [answer, score, strt_token , End_token]
      df.loc[len(df)] = list_row









  # Sort Result
  df.to_excel("output.xlsx")
  df=df.sort_values(by=['score'],ascending = False)
  df.to_excel("output1.xlsx")

  # post_processing

  df2 = pd.DataFrame(columns = ['answer', 'score', 'strt_token_indx', 'end_token_indx'])
  ptexta=Dict_Passages[pid1]
  ptexta=ptexta.replace("\"","")
  passage_list=ptexta.split()
  mask=[0]* len(passage_list)

  for index, row in df.iterrows():
    anspo=str(row['answer']).strip()
    anspo=anspo.replace("\"","")
    scopo=str(row['score']).strip()
    strtpo=str(row['strt_token_indx']).strip()
    endpo=str(row['end_token_indx']).strip()
    ans_list=anspo.split()


    worda=anspo


    if stp.is_stop(worda)==False:
      start_mask=-1
      End=-1
      Found=False

      for index_list in range(len(passage_list)-len(ans_list)+1):
        start_mask=index_list
        End=index_list+len(ans_list)
        if passage_list[start_mask:End]==ans_list:
          Found=True
          break


      if Found==False:
        for index_list in range(len(passage_list)-len(ans_list)+1):
          start_mask=index_list+1
          End=index_list+len(ans_list)
          if passage_list[start_mask:End]==ans_list[1:]:
            start_mask=index_list
            Found=True
            break

      if Found==False:
        for index_list in range(len(passage_list)-len(ans_list)+1):
          start_mask=index_list
          End=index_list+len(ans_list)-1
          if passage_list[start_mask:End]==ans_list[:len(ans_list)-1]:
            End=index_list+len(ans_list)
            Found=True
            break

      if Found==False:
        for index_list in range(len(passage_list)-len(ans_list)+1):
          start_mask=index_list+1
          End=index_list+len(ans_list)-1
          if passage_list[start_mask:End]==ans_list[1:len(ans_list)-1]:
            start_mask=index_list
            End=index_list+len(ans_list)
            Found=True
            break


      Ans_mask=mask[start_mask:End]
      index_of_one=[]
      seq=[]
      seq_end=[]


      for check_index in range(len(Ans_mask)):
        if Ans_mask[check_index]==0:
          index_of_one.append(check_index)

      if len(Ans_mask)==len(index_of_one):
        for chang in range(start_mask,End):
            mask[chang]=1
        list_row = [anspo, scopo, strtpo , endpo]
        df2.loc[len(df2)] = list_row


      elif len(index_of_one)> 0:
        seq_check=index_of_one[0]
        seq.append(index_of_one[0])

        for seq_indx in range(1,len(index_of_one)):
          if index_of_one[seq_indx] !=  seq_check+1:
            seq.append(index_of_one[seq_indx])
            seq_end.append(index_of_one[seq_indx-1])
            seq_check=index_of_one[seq_indx]
          else:
            seq_check=index_of_one[seq_indx]

        seq_end.append(index_of_one[len(index_of_one)-1])

        for seq_list in range(len(seq_end)):
          s2=seq[seq_list]
          e2=seq_end[seq_list]

          new_seq=ans_list[s2:e2+1]

          start_mask2=-1
          End22=-1
          Found2=False
          for index_list in range(len(passage_list)-len(new_seq)+1):
            start_mask2=index_list
            End22=index_list+len(new_seq)
            if passage_list[start_mask2:End22]==new_seq:
              Found2=True
              break


          if new_seq !=['.']:
            strtpo=start_mask+s2
            endpo=start_mask+e2+1-1
            list_row = [new_seq, scopo, start_mask2 , End22-1]
            df2.loc[len(df2)] = list_row
            for chang2 in range(start_mask2,End22):
              mask[chang2]=1



  df=df2

  # post_processing end


  # Select the top 10


  emptyFlage=False
  if len(df)==0:
    emptyFlage=True
  elif len(df)<10:
    new_df=df
  else:
    new_df=df.head(10)




  dic_text=""
  if emptyFlage==False:
    first_answer=True
    AC=1
    for index, row in new_df.iterrows():
      if first_answer==True:
        first_answer=False
        dic_text+="{\"answer\":"+"\""+str(row['answer']).strip()+"\""+ ",\"rank\":"+ str(AC)+ ",\"score\":"+ str(row['score'])+",\"strt_token_indx\":"+ str(row['strt_token_indx'])+",\"end_token_indx\":"+ str(row['end_token_indx']) +"}"
      else:
        dic_text+=",{\"answer\":"+"\""+str(row['answer']).strip()+"\""+ ",\"rank\":"+ str(AC)+ ",\"score\":"+ str(row['score'])+",\"strt_token_indx\":"+ str(row['strt_token_indx'])+",\"end_token_indx\":"+ str(row['end_token_indx']) +"}"
      AC+=1


  if len(answers_list)>0:
    if fpid==False:
      ALL_STRING_REsult+=pid1+":"+"["+dic_text+"]"+"\n"
      fpid=True
    else:
      ALL_STRING_REsult+=","+pid1+":"+"["+dic_text+"]"+"\n"
  else:
    if fpid==False:
      ALL_STRING_REsult+=pid1+": []"+"\n"
      fpid=True
    else:
      ALL_STRING_REsult+=","+pid1+": []"+"\n"



ALL_STRING_REsult=ALL_STRING_REsult.strip()+"}"

save_path="/content/sample_data/output/LKAU23_run02.json"
with open(save_path, 'w', encoding="utf-8") as f:
  f.write(ALL_STRING_REsult)
```
